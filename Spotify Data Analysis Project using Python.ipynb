```python
# Converting duration in milliseconds to just seconds
# We have used lambda function and dividing the values by 1000 so its converted into seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
# Create first visualization that is going to be a correlation map
# We are also going to drop 3 unwanted columns key, mode and explicit and then we are going
# to apply Pearson correlation method and we haveset fig size 14, 6

# Then, we are using seaborn heatmap to create our Correlation map, 
# annot = True - will write the data value in each cell
# fmt = .1g - string formatting code for annotation
# vmin vmax - to anchor the colourmap otherwise inferred from data and other keyboard arguments
# cmap - colour map

corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidth = 1, linecolor = "Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
# Select only numeric columns for correlation
numeric_df = df_tracks.select_dtypes(include=[float, int])

# Compute the correlation matrix
corr_df = numeric_df.corr(method="pearson")

# Plot the heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

On right side scale ranges from -1 to +1
-1 means variable have negative correlation and vice-versa
We have values like -0.7 for Energy and Acousticness, which means if Energy is high Acoustic will be low
Again for Loudness and Speechness, if we the song is loud speechness is low which is negative correlation
But if we see for Energy and Loudness there is really high correlation variable (0.8)
There is -ve corelation bw Acousticness and Danceability
-ve correlation bw Valence and Acousticness
+ve bw Energy and Valence(0.4)

```python
# Sample a data and going to take 0.4% of data and create 2 Regression plots

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
# Create a Regression plot bw Loudness and Energy, in Correlation map we saw there was +ve 
# correlation bw loudness and energy (0.8) let's plot in form of Regression line.

plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

We can find there is very high correlation bw energy and loudness. We can find all the datapoints for the songs are in one direction so, if Energy increases Lodness of the song also increases and vice-versa.

```python
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "c").set(title = "Popularity vs Acousticness Correlation")
```

We can find regression line is downwards which means as Acousticness of song increases Popularity decreases and vice-versa.

```python
# 2nd Data Analysis Project based on Genres of the Songs

df_genre = pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.barplot(x='duration_ms', y='genre', data=df_genre, hue='genre', palette="rocket", legend=False)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
plt.show()
```

For Classical and World Genre duration of songs are more compared to others like Children's Music the duration is least.

```python
# Top 5 Genres by Popularity
# We took 10 because few genres are repetitive

sns.set_style(style="darkgrid")
plt.figure(figsize=(10, 5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y="genre", x="popularity", data=famous).set(title="Top 5 Genres by Popularity")
```
